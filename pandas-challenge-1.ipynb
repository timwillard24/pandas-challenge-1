## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
pandas_challenge_df['subtotal'] = pandas_challenge_df['qty'] * pandas_challenge_df['unit_price']
pandas_challenge_df.columns
```

```python
pandas_challenge_df['shipping_weight'] = pandas_challenge_df['qty'] * pandas_challenge_df['unit_weight']
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def calc_shipping_price(shipping_weight):
    if shipping_weight > 50:
        return shipping_weight * 7 
    else:
        return shipping_weight * 10
pandas_challenge_df['total_shipping_price'] = pandas_challenge_df ['shipping_weight'].apply(calc_shipping_price)
pandas_challenge_df.columns
pandas_challenge_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
pandas_challenge_df['total_price'] = (pandas_challenge_df['subtotal'] + pandas_challenge_df['total_shipping_price']) * 1.0925
pandas_challenge_df.columns
pandas_challenge_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

pandas_challenge_df['line_cost'] = (pandas_challenge_df['unit_cost'] * pandas_challenge_df ['qty']) + pandas_challenge_df ['total_shipping_price']
pandas_challenge_df.columns
pandas_challenge_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price

pandas_challenge_df['profit'] = pandas_challenge_df['total_price'] - pandas_challenge_df ['line_cost']
pandas_challenge_df.columns
pandas_challenge_df.head()
```

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top5_client_ids = [33615, 66037, 46820, 38378, 24741]
top5_clients_df = pandas_challenge_df[pandas_challenge_df['client_id'].isin(top5_client_ids )]
top5_clients_spending = top5_clients_df.groupby('client_id')['total_price'].sum().round(2)
print(top5_clients_spending)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# Assuming top5_client_ids contains the IDs of the top 5 clients
top5_clients_summary_df = pandas_challenge_df[pandas_challenge_df['client_id'].isin(top5_client_ids)]

top5_clients_summary_df = top5_clients_summary_df.groupby('client_id').agg({
    'qty': 'sum',  # total units purchased
    'total_shipping_price': 'sum',
    'total_price': 'sum',  # total revenue
    'profit': 'sum'  # total profit
}).reset_index()

# Rename the columns for clarity
top5_clients_summary = top5_clients_summary.rename(columns={
    'qty': 'Total Units Purchased',
    'total_shipping_price': 'Total Shipping Price',
    'total_price': 'Total Revenue',
    'profit': 'Total Profit'
})

# Display the summary DataFrame
top5_clients_summary_df


```

```python
# Define the money columns
money_columns = ['total_shipping_price', 'total_price', 'profit']

# Define a function to convert a dollar amount to millions
def convert_to_millions(amount):
    return '${:.6f}'.format(amount / 1000000)

# Apply the conversion function to the money columns
for col in money_columns:
    top5_clients_summary_df[col] = top5_clients_summary_df[col].apply(convert_to_millions)

# Rename the columns to reflect the change in the money format
top5_clients_summary_df = top5_clients_summary_df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
    'total_shipping_price': 'Shipping (millions)',
    'total_price': 'Total Revenue (millions)',
    'profit': 'Total Profit (millions)'
})

# Display the formatted DataFrame
top5_clients_summary_df
```

```python

top5_clients_summary_df = top5_clients_summary_df.sort_values(by='Total Profit (millions)', ascending=False)
top5_clients_summary_df
```

```python
# Convert the columns to numeric values
top5_clients_summary_df['Shipping (millions)'] = top5_clients_summary_df['Shipping (millions)'].str.replace('$', '').str.replace('M', '').astype(float)
top5_clients_summary_df['Total Revenue (millions)'] = top5_clients_summary_df['Total Revenue (millions)'].str.replace('$', '').str.replace('M', '').astype(float)
top5_clients_summary_df['Total Profit (millions)'] = top5_clients_summary_df['Total Profit (millions)'].str.replace('$', '').str.replace('M', '').astype(float)

# Apply the formatting
top5_clients_summary_df['Shipping (millions)'] = top5_clients_summary_df['Shipping (millions)'].apply(lambda x: '${:.2f}M'.format(x))
top5_clients_summary_df['Total Revenue (millions)'] = top5_clients_summary_df['Total Revenue (millions)'].apply(lambda x: '${:.2f}M'.format(x))
top5_clients_summary_df['Total Profit (millions)'] = top5_clients_summary_df['Total Profit (millions)'].apply(lambda x: '${:.2f}M'.format(x))

# Display the formatted DataFrame
top5_clients_summary_df
```
